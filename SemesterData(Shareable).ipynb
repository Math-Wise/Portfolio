```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load data and print head
df = pd.read_csv('/content/drive/MyDrive/ORCA/Semester 1 Grades/Shareable/Semester1Grades(Shareable).csv')
df.head()
```

## Distributions

```python
#Overall semester 1 grades distribution
passing_count = 0
total_count = len(df['Semester 1 Grade'])

for score in df['Semester 1 Grade']:
  if score >= 60:
    passing_count += 1

sns.set_style("whitegrid")

sns.histplot(data = df, x = 'Semester 1 Grade', binrange = (0, 100), binwidth = 5)
plt.axvspan(60, 100, color = 'green', alpha = 0.2)
plt.text(61, 14.5, f"{passing_count} out of {total_count} passed.", fontsize = 12)
plt.title('Semester 1 Grade Distribution', fontsize = 18)
plt.xlabel('Scores (%)', fontsize = 12)
plt.xticks(range(0, 101, 10))
plt.ylabel('Counts', fontsize = 12);
```

```python
#Exam bucket distribution
sns.histplot(data = df, x = 'Exam Total', binrange = (0, 100), binwidth = 5)
plt.title('Exam Score Distribution (75% of Grade)', fontsize = 18)
plt.xlabel('Exam Scores', fontsize = 12)
plt.xticks(range(0, 101, 10))
plt.ylabel('Counts', fontsize = 12);
```

```python
#Exercises assessment bucket distribution
sns.histplot(data = df, x = 'EA Total', binrange = (0, 100), binwidth = 5)
plt.title('Exercise Assesment Distribution (10% of Grade)', fontsize = 18)
plt.xlabel('Exercise Assesment Scores', fontsize = 12)
plt.xticks(range(0, 101, 10))
plt.ylabel('Counts', fontsize = 12)
plt.yticks(range(0,46, 5));
```

```python
#Homework bucket distribution
sns.histplot(data = df, x = 'HW Total', binrange = (0, 100), binwidth = 5)
plt.title('Homework Distribution (8% of Grade)', fontsize = 18)
plt.xlabel('Homework Scores', fontsize = 12)
plt.xticks(range(0, 101, 10))
plt.ylabel('Counts', fontsize = 12)
plt.yticks(range(0,51,5));
```

```python
#Learn startegies bucket distribution
sns.histplot(data = df, x = 'LS Total', binrange = (0, 100), binwidth = 5)
plt.title('Learning Strategies Distribution (5% of Grade)', fontsize = 18)
plt.xlabel('Learning Strategies Score', fontsize = 12)
plt.xticks(range(0, 101, 10))
plt.ylabel('Counts', fontsize = 12);
```

## Scores Throught the Semester

```python
#Scores for tests
test_scores = {
    'Exam 1': df['CA Exam 1'].mean(),
    'Exam 2': df['CA Exam 2'].mean(),
    'Exam 3': df['CA Exam 3'].mean(),
    'Midterm': df['CA Semester 1 Midterm'].mean()
}


plt.plot(test_scores.values())

plt.title('Mean Test Scores', fontsize = 18)
plt.ylabel('Exam Mean (%)', fontsize =12)
plt.yticks(range(0, 51, 5))
plt.xticks(range(4), test_scores, rotation = 45, fontsize = 12);
```

```python
#Scores for exercise assessments
exercise_assessment_scores = {
    'EA Unit 1 & 2': df['EA: Unit 1 & 2'].mean(),
    'EA Unit 3 & 4': df['EA Unit 3 & 4'].mean(),
    'EA Unit 5 & 6': df['EA Unit 5 & 6'].mean()
}

plt.plot(exercise_assessment_scores.values())

plt.title('Exercise Assessment Scores', fontsize = 18)
plt.ylabel('Mean Score (%)', fontsize =12)
plt.yticks(range(0, 36, 5))
plt.xticks(range(3), exercise_assessment_scores, rotation = 45, fontsize = 12);
```

```python
#Scores for homework
homework_scores = {
    "CW Unit 1" : df['CW Unit 1'].mean(),
    "CW Unit 2 Part 1" : df['CW Unit 2 Part 1'].mean(),
    "CW Unit 2 Part 2" : df['CW Unit 2 Part 2'].mean(),
    "CW Unit 2 Part 3" : df['CW Unit 2 Part 3'].mean(),
    "CW Unit 3" : df['CW Unit 3'].mean(),
    "CW Unit 4 Part 1" : df['CW Unit 4 Part 1'].mean(),
    "CW Unit 4 Part 2" : df['CW Unit 4 Part 2'].mean(),
    "CW Unit 4 Part 3" : df['CW Unit 4 Part 3'].mean(),
    "CW Unit 5 Part 1" : df['CW Unit 5 Part 1'].mean(),
    "CW Unit 5 Part 2" : df['CW Unit 5 Part 2'].mean(),
    "CW Unit 5 Part 3" : df['CW Unit 5 Part 3'].mean(),
    "CW Unit 6" : df['CW Unit 6'].mean()
}

plt.plot(homework_scores.values())

plt.title('Homework Completion', fontsize = 18)
plt.ylabel('Percent Turned In', fontsize = 12)

plt.xticks(range(12), homework_scores, rotation = 90)
plt.yticks(np.array(range(0, 101, 10))/100);
```

```python
##Scores for Learning Startegies
learning_startegies_scores = {
    'LS 1 Quiz': df['LS 1 Quiz'].mean(),
    'LS 2 Quiz': df['LS2 Quiz'].mean()
}

plt.plot(learning_startegies_scores.values())

plt.title('Learning Strategies Scores', fontsize = 18)
plt.ylabel('Mean Score (%)', fontsize =12)
plt.yticks(range(0, 71, 10))
plt.xticks(range(2), learning_startegies_scores, rotation = 45, fontsize = 12);
```
